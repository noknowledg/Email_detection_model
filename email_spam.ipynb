```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(r"C:\Users\shukl\Downloads\spam.csv")
df.head(10)
```

```python
df.groupby('Category').describe()
```

```python
df['spam'] = df['Category'].apply(lambda x:1 if x== 'spam' else 0)
df.head()
```

```python
x = df.Message
y = df.spam
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=123)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
```

```python
x_train_encoded = vectorizer.fit_transform(x_train)
x_test_encoded = vectorizer.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
x_train_encoded = x_train_encoded.toarray()
x_test_encoded= x_test_encoded.toarray()
```

```python
multiNominal = MultinomialNB()
```

```python
multiNominal.fit(x_train_encoded,y_train)
```

```python
y_pred  = multiNominal.predict(x_test_encoded)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

 Sklearn Pipelearn

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
  ('Vectorizer', CountVectorizer()),
('nb', MultinomialNB())
 ])
```

```python
clf.fit(x_train, y_train)
```

```python
clf.score(x_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

```python
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(),
    'KNN': KNeighborsClassifier(n_neighbors=7),
    'SVM': SVC(kernel='linear'),
    'Naive Bayes': GaussianNB()
}
```

```python
for name, model in models.items():
    model.fit(x_train_encoded,y_train)
    y_pred = model.predict(x_test_encoded)

    print(f"Accuracy of {name} : {accuracy_score(y_test,y_pred)}")
    print(f'{confusion_matrix(y_test,y_pred)}')
```
